## Task 2:
Loading and Visualizing Images

```python
# a. Loading/reading image using OpenCv
import cv2
img = cv2.imread(r"/content/A-bouquet-of-roses-1895.jpg") #Load the image
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()

```

```python
# b. Loading/reading image using PIL
from PIL import Image
img2 = Image.open(r"/content/The-Watzmann.jpg") #load the image using PIL
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.imshow(img2, cmap = cm.Greys_r)
plt.show()
```
